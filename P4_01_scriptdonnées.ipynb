```python
#jointure entre les dataframes ventes et produits
ventes_produits = pd.merge(ventes, produits, on= 'id_prod', how= 'left')
ventes_produits
```

```python
# jointure entre les dataframes ventes_produits et clients
data = pd.merge(ventes_produits, clients, on= 'client_id', how= 'right')

# dataframe classé dans l'ordre croissant selon la colonne 'price' afin de rechercher d'éventuels prix <= 0
data.sort_values('price', ascending=True)
```

```python
# supression des sessions de test
data.drop(data.loc[data['session_id']== 's_0'].index, inplace=True)

# vérification du nombre de valeurs nulles
data.isnull().sum()
```

```python
# affichage de tous les index du dataframe "data" contenant des NaN
valeurs_manquantes = data[data['price'].isnull() == True]
valeurs_manquantes
```

```python
# recherche du ou des produit(s) n'ayant pas de prix
valeurs_manquantes['id_prod'].unique()
```

```python
# il semble s'agir d'un produit de la catégorie 0
# les produits de la catégorie 0 semble être nommés 0_XXXX
# création d'une variable contenant tous les achats concernants au moins un produit de la catégorie 0
categ_0 = data[data['categ']== 0]

# on remplace les valeurs NaN de la coolonne 'price' par la moyenne des prix de la catégorie 0
data['price'] = data['price'].fillna(categ_0['price'].mean())

# les valeurs manquantes de la colonne 'categ' sont remplacées par 0
data['categ'] = data['categ'].fillna(0)
```

```python
data[data['date'].isnull() == True]
```

```python
# les index du dataframe 'data' soont supprimés faute de pouvoir les remplacer de manière cohérente
data = data.dropna()
```

```python
# on change le format object de la colonne 'date' pour le format datetime
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d %H:%M:%S', errors= 'coerce')
```

```python
# création de la colonne 'mois' pour pouvoir créer des graphiques périodiquement mensuels
data['mois'] = data['date'].apply(lambda x: x.strftime('%b %Y'))
```

```python
# création d'une colonne précisant l'âge des clients
data['age'] = 2022- data['birth']

# création de tranches d'âges
cut_bins = [17, 29, 45, 59, 99]
cut_labels = ['18-29', '30-45', '46-59', '60+']
data['tranche'] = pd.cut(data['age'], bins=cut_bins, labels=cut_labels)
```
